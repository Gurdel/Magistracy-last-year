```python
class Node:
    def __init__(self, freq, value, left=None, right=None):
        self.freq = freq
        self.value = value
        self.left = left
        self.right = right
        self.code = ""
```

```python
def calculate_frequencies(data):
    frequencies = dict()
    for element in data:
        if frequencies.get(element) == None:
            frequencies[element] = 1
        else:
            frequencies[element] += 1
    
    return frequencies
```

```python
codes = dict()

def init_dictionary(heap, val=""):
    new_val = val + str(heap.code)

    if heap.left:
        init_dictionary(heap.left, new_val)
    if heap.right:
        init_dictionary(heap.right, new_val)
    if not heap.left and not heap.right:
        codes[heap.value] = new_val

    return codes
```

```python
def apply_coding(data, coding):
    encoding_output = []
    for s in data:
        encoding_output.append(coding[s])

    string = "".join([str(item) for item in encoding_output])
    return string
```

```python
def encode(data):
    frequencies = calculate_frequencies(data)

    nodes = []
    for value, freq in frequencies.items():
        nodes.append(Node(freq, value))

    while len(nodes) > 1:
        nodes = sorted(nodes, key=lambda x: x.freq)

        right = nodes[0]
        left = nodes[1]

        left.code = 0
        right.code = 1

        newNode = Node(left.freq + right.freq, left.value + right.value, left, right)

        nodes.remove(left)
        nodes.remove(right)
        nodes.append(newNode)

    dictionary = init_dictionary(nodes[0])

    encoded_output = apply_coding(data, dictionary)

    return (
        encoded_output,
        nodes[0],
        dictionary,
        frequencies,
    )
```

```python
from scipy.io import wavfile

VISUAL_LENGTH = 200

fs, original_signal = wavfile.read('./file.wav')

#  Stereo
original_signal = [s[0] for s in original_signal]
print('Orignal:', original_signal[:VISUAL_LENGTH])

# #  Mono
# original_signal = list(original_signal)
# print('Orignal:', original_signal[:VISUAL_LENGTH])

(
    encoding,
    tree_heap,
    dictionary,
    frequencies
) = encode(original_signal)

print('Encoded:', encoding[:VISUAL_LENGTH])
```

```python
def decode(encoded_data, huffman_tree):
    tree_head = huffman_tree
    decoded_output = []
    for x in encoded_data:
        if x == " ":
            continue
        if x == "1":
            huffman_tree = huffman_tree.right
        elif x == "0":
            huffman_tree = huffman_tree.left
        try:
            if huffman_tree.left.value == None and huffman_tree.right.value == None:
                pass
        except AttributeError:
            decoded_output.append(huffman_tree.value)
            huffman_tree = tree_head

    string = " ".join([str(item) for item in decoded_output])
    return list(map(int, string.split()))
```
