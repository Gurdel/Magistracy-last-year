# Large dataset generation

```python
import os

data_folder_path = os.path.join('.', 'data')
books_list = os.listdir(data_folder_path)
books_list
```

```python
import codecs
import re

words = []
for book in books_list:
    book_path = os.path.join(data_folder_path, book)
    with codecs.open(book_path, "r", "utf_8_sig") as fileObj:
        text = fileObj.read()

    words += set(re.findall("[a-z]+[’'-]?[a-z]*", text.lower()))
print('\tTotal words count:\t', len(words))
```

```python
import random

UNIQ = '2'
for i in range(3000):
    with open(f'./generated/g{UNIQ}_{i}.txt', 'w') as f:
        for j in range(random.randint(2_000, 15_000)):
            f.write(' ' .join(random.choices(words, k=random.randint(5, 150)))+'\n')
```
